```python
import yaml as ym
from yaml import Loader, load, Dumper, dump
import pandas as pd
```

```python
file = open("composer_eras.yml","r").read()
data = ym.load(file,Loader=Loader)

l = []
for m in data.keys():
   for x in data[m]:
      i = [x,m]
      l.append(i)

df = pd.DataFrame(l,columns=['Composer','Period'])
df
```
