```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
graph = nx.Graph()
graph.add_edges_from([
   ("MI","Josh"),
   ("MI","Brittanie"),
   ("7C","Trisha"),
   ("7C","Tiffany"),
   ("S","Josh"),
   ("S","Trisha"),
   ("MI","Tiffany"),
   ("7C","Josh"),
   ("S","Brittanie"),
   ("MI","Trisha")])
```

```python
options = {"node_size" : 3000,
           "node_color" : "white",
           "edgecolors" : "black",
           "linewidths" : 8,
           "width" : 5}
```

```python
nx.draw(graph, **options,with_labels=True)

ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

```python
G = nx.Graph()

G.add_node("a",node_color="green")
G.add_edge("a", "b", weight=0.6)
G.add_edge("a", "c", weight=0.2)
G.add_edge("c", "d", weight=0.1)
G.add_edge("c", "e", weight=0.7)
G.add_edge("c", "f", weight=0.9)
G.add_edge("a", "d", weight=0.3)


elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] > 0.5]
esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] <= 0.5]

pos = nx.spiral_layout(G)  # positions for all nodes - seed for reproducibility

# nodes
nx.draw_networkx_nodes(G, pos, node_size=700, node_color=["green","red","blue","orange","white","yellow"])

# edges
nx.draw_networkx_edges(G, pos, edgelist=elarge, width=6, edge_color = 'white')
nx.draw_networkx_edges(
    G, pos, edgelist=esmall, width=2, alpha=0.5, edge_color="white", style="dashed")
# labels
nx.draw_networkx_labels(G, pos, font_size=10, font_family="sans-serif")

ax = plt.gca()
ax.margins(0.08)
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
mg = nx.Graph()
drugs = "pembrolizumab nivolumab bevacizumab tocilizumab".split(" ")
mg.add_nodes_from(drugs,size=100)
mg.add_edges_from([(drugs[0],drugs[1]),
                   (drugs[0],drugs[3]),
                   (drugs[0],drugs[2]),
                   (drugs[2],drugs[3])])
nx.draw(mg, with_labels=True,font_size=10)
ax = plt.gca()
ax.margins(0.08)
plt.axis("off")
plt.show()
```
